```python
#%pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def p_log_p(p):
    return p * np.log2(p, where=np.not_equal(p, 0))

def entropy(p):
    return ((-1) * p_log_p(p)) + ((-1) * p_log_p(1 - p))

def plot_entropy():
    p = np.linspace(0, 1, num=1000)
    h_p = entropy(p)
    plt.plot(p, h_p)
    plt.xlabel('p')
    plt.ylabel('Entropy of X_p')
    plt.title('Entropy of a random variable')
    # mark the maximum
    max_index = np.argmax(h_p)
    max_p = p[max_index]
    plt.plot(max_p, entropy(max_p), 'ro')
    plt.annotate(f'Max: p = {max_p:.2f}; H(X_p) = {entropy(max_p):.2f}', (max_p, entropy(max_p)))
    plt.show()

plot_entropy()
```
